### There are 3 main types of Machine Learning algorithms：
#### Supervised Learning：

You have labeled data and are trying to predict a label based off of known features.

    1. Such as an input or the desired output is known.
    
    2. Receives a set of inputs along with the corresponding correct outputs and the algorithm learns by comparing its actual output with correct outputs to find errors.
    
    3. Then modifies the model accordingly.
    
    4. More methods like classfication,regression,prediction and gradient boosting.
    
    5. Use patterns to predict the values of the label on additional unlabeled data.
    
    6. Commonly used in application where historical data predicts likely future events.

#### Unspervised Learning：
_____
You have unlabeled data and trying to group together similar data points based off of features.

    1. It is used against data that has no historical labels.
    
    2. The system is not told the "right answer", it must figure out what is being shown.
    
    3. The goal is to explore the data and find some sort of structure within the data.
    
    4. Can find the main attributes that separate customer segments from each other.
    
    5. More methods like self-organizing maps,nearest-neighbor mapping,k-means clustering and singular value decomposition.
    
    6. These methods are used to segment text topics,recommend items and identify data outliers.
_____    
#### Reinforcement Learning：

Algorithms learns to perform an action based off of previous experience.

    1. Often used for robotics gaming and navigation.
    
    2. The algorithm discovers through trial and error which actions yield the greatest rewards.
    
    3. This type of learning has three primary components：
    
        1. The agent(the learner or decision maker).
        2. The environment(everything the agent interacts with).
        3. Actions(what the agent can actually do or perform).
    
    4. The object is for the agent to choose actions that max the expected reward over a given amount of time.
    
    5. The agent will reach the goal much faster by following a good policy.
    
    6. The goal is to learn the best policy.
